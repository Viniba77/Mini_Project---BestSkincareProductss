# Analysing and visualizing Skin Care Products

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('cosmetic.csv')
df.tail()
```

```python
df.describe()
```

```python
df
```

```python
df.Label.unique()
```

```python
plt.figure(figsize=(15,10))
df.Label.value_counts(dropna = False).plot(kind = 'pie',autopct='%1.0f%%')
plt.title('Proportion of Products',fontsize = 25)
plt.show()
```

```python
df_best =df[['Brand','Rank','Label','Price']]
df_best
```

```python
df_best.set_index('Brand',inplace =True)
```

```python
df_best.sort_values(by ='Rank',ascending =False).head(10).plot(kind ='bar')
plt.show()
```

```python
df['Average_Price'] = df.Price.mean()
```

```python
df
```

```python
df_best =df[['Brand','Rank','Label']]
df_best.set_index('Brand',inplace = True)
```

```python
df_best.loc[df_best.Rank >=5].sort_values(by = 'Rank')
```

# Best Moisturizer (10)

```python
df_best_moist=df_best[df_best.Label=='Moisturizer'].sort_values(by='Rank',ascending =False).head(10)
df_best_moist.plot(kind = 'bar')
plt.ylabel('Rank')
plt.show()
```

# Water based Moisturizer

```python
water = df.Ingredients.str.contains('Water') 
water
```

```python
moi = df.Label=='Moisturizer'
moi
```

```python
import matplotlib.pyplot as plt
df.loc[water & moi,['Label','Brand','Rank','Ingredients']].sort_values(by='Rank',ascending=False)
```

```python
df.loc[water & moi,['Label','Brand','Rank','Ingredients']].sort_values(by='Rank',ascending=False).set_index('Brand').head(10).plot(kind='bar')
plt.xlabel('Brand')
plt.ylabel('Rank')
plt.show()
```

# Budget Friendly Sun protect

```python
df.Ingredients[1]
```

```python
df
```

```python
plt.figure(figsize=(25,12))
su = df.loc[(df.Label=='Sun protect') & (df.Rank >4),['Label','Brand','Rank','Price','Ingredients']].set_index('Brand').sort_values(by= 'Rank',ascending= True).head(10).plot(kind ='bar')
plt.xlabel('Brand',fontsize = 15)
plt.title('Budget Friendly Sunprotect',fontsize = 20)
plt.ylabel('Price',fontsize = 15)
plt.show()
```

# Budget Friendly Facemask

```python
plt.figure(figsize=(25,12))
su = df.loc[(df.Label=='Face Mask') & (df.Rank >4),['Label','Brand','Rank','Price','Ingredients']].set_index('Brand').sort_values(by= 'Rank',ascending= True).head(10).plot(kind ='bar')
plt.xlabel('Brand',fontsize = 15)
plt.title('Budget Friendly Face Mask',fontsize = 20)
plt.ylabel('Price',fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(25,12))
su = df.loc[(df.Combination==1) & (df.Normal ==1)& (df.Sensitive == 1)& (df.Rank >=4),['Label','Brand','Rank','Price','Ingredients']].set_index('Brand').sort_values(by= 'Rank',ascending= True).head(10).plot(kind ='bar',color='g')
plt.xlabel('Brand',fontsize = 15)
plt.title('Combination Products',fontsize = 20)
plt.ylabel('Price',fontsize = 15)
plt.show()
```
